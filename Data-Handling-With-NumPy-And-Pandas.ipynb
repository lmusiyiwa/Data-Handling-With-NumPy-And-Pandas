```python
import numpy as np
import matplotlib.pyplot as plt

# --- Assignment 1: WEATHER SENSOR DATA ANALYSIS ---

weather_data = np.array([
    [[30, 70], [32, 65], [28, 80], [31, 75]],   # Lagos
    [[25, 60], [27, 55], [29, 50], [26, 58]],   # Abuja
    [[33, 40], [35, 45], [34, 42], [36, 38]]    # Kano
])

# --- Exercise 1: Shape, size, dimensions
print("Shape:", weather_data.shape)
print("Size:", weather_data.size)
print("Dimensions:", weather_data.ndim)

# --- Exercise 2: Extract values
lagos_temps = weather_data[0,:,0]
print("Lagos Temperatures:", lagos_temps)

day2_humidity = weather_data[:,1,1]
print("Day 2 Humidity across cities:", day2_humidity)

kano_day3 = weather_data[2,2,:]
print("Kano Day 3:", kano_day3)

# --- Exercise 3: Specific temps
temps = weather_data[[0,1,2],[0,1,2],0]
print("Selected Temps (Lagos Day1, Abuja Day2, Kano Day3):", temps)

# --- Exercise 4: Conditions
days_temp_gt30 = np.where(weather_data[:,:,0] > 30)
print("Days with Temp > 30°C:", days_temp_gt30)

cities_humidity_lt50 = np.where(weather_data[:,:,1] < 50)
print("Cities with Humidity < 50%:", cities_humidity_lt50)

# --- Exercise 5: Stack temp & humidity
temps_only = weather_data[:,:,0]
humidity_only = weather_data[:,:,1]
print("Temperature Array:\n", temps_only)
print("Humidity Array:\n", humidity_only)

# --- Exercise 6: Sorting & max
sorted_temps = np.sort(weather_data[:,:,0].flatten())
print("Sorted Temperatures:", sorted_temps)

abuja_highest_day = np.argmax(weather_data[1,:,0])
print("Day with Highest Temp in Abuja:", abuja_highest_day+1)

# --- Exercise 7: Average temp per city
avg_temp_city = np.mean(weather_data[:,:,0], axis=1)
print("Average Temp per City:", avg_temp_city)

# --- Exercise 8: City with lowest humidity
lowest_humidity_city = np.argmin(np.mean(weather_data[:,:,1], axis=1))
print("City with Lowest Humidity (0=Lagos,1=Abuja,2=Kano):", lowest_humidity_city)

# --- Assignment 2: PATIENT VITALS MONITORING SYSTEM ---

np.random.seed(0)
patients_data = np.zeros((4,3,2,4))

patients_data[:,:,:,0] = np.random.randint(60,101,(4,3,2))   # Heart Rate
patients_data[:,:,:,1] = np.random.randint(90,121,(4,3,2))   # SBP
patients_data[:,:,:,2] = np.random.randint(60,81,(4,3,2))    # DBP
patients_data[:,:,:,3] = np.random.uniform(36,39,(4,3,2))    # Temp

# --- Exercise 1: Patient 0, Day 3
patient0_day3 = patients_data[0,2,:,:]
print("Patient 0, Day 3 Vitals:\n", patient0_day3)

# --- Exercise 2: Patient 3 Temperatures
patient3_temps = patients_data[3,:,:,3]
print("Patient 3 Temperatures:\n", patient3_temps)

# --- Exercise 3: Evening HR for all patients, Day 2
evening_hr_day2 = patients_data[:,1,1,0]
print("Evening HR Day 2 (all patients):", evening_hr_day2)

# --- Exercise 4: Patients 1,2,4 only
subset_patients = patients_data[[0,1,3],:,:,:]
print("Subset Patients (1,2,4): Shape", subset_patients.shape)

# --- Exercise 5: Days 2,3 for Patient 2
patient2_days23 = patients_data[1,1:3,:,:]
print("Patient 2, Days 2 & 3:\n", patient2_days23)

# --- Exercise 6: Temp > 37.5
high_temps = np.where(patients_data[:,:,:,3] > 37.5)
print("Indices where Temp > 37.5°C:", high_temps)

# --- Exercise 7: Patients with SBP > 140
bp_high_patients = np.unique(np.where(patients_data[:,:,:,1] > 140)[0])
print("Patients with SBP > 140:", bp_high_patients)

# --- Exercise 8: Average HR per patient
avg_hr = np.mean(patients_data[:,:,:,0], axis=(1,2))
print("Average HR per Patient:", avg_hr)

# --- Exercise 9: Max Temp per day for Patient 3
max_temp_patient3 = np.max(patients_data[3,:,:,3], axis=1)
print("Max Temp per Day for Patient 3:", max_temp_patient3)

# --- Exercise 10: Overall mean SBP
mean_sbp = np.mean(patients_data[:,:,:,1])
print("Overall Mean SBP:", mean_sbp)

# --- Exercise 11: Stack HR & Temp only
hr_temp = patients_data[:,:,:,[0,3]]
print("HR & Temp Array Shape:", hr_temp.shape)

# --- Exercise 12: Sort avg HR ascending
sorted_avg_hr = np.sort(avg_hr)
print("Sorted Avg HR:", sorted_avg_hr)

# --- Visualizations ---

# Bar chart: Average Temperature per City
cities = ["Lagos", "Abuja", "Kano"]
plt.bar(cities, avg_temp_city, color=['orange','green','red'])
plt.title("Average Temperature per City")
plt.ylabel("Temperature (°C)")
plt.show()

# Line chart: Average Heart Rate per Patient
patients = ["Patient 1","Patient 2","Patient 3","Patient 4"]
plt.plot(patients, avg_hr, marker='o')
plt.title("Average Heart Rate per Patient")
plt.ylabel("Heart Rate (bpm)")
plt.show()
```
